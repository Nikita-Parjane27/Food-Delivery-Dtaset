```python
## Load CSV

import pandas as pd

orders_df = pd.read_csv("orders.csv")
```

```python
##Load JSON Data

users_df = pd.read_json("users.json")
```

```python
import sqlite3

conn = sqlite3.connect("restaurants.db")
cursor = conn.cursor()

with open("restaurants.sql", "r") as file:
    sql_script = file.read()

cursor.executescript(sql_script)
conn.commit()
```

```python
restaurants_df = pd.read_sql(
    "SELECT * FROM restaurants",
    conn
)

restaurants_df.head()
```

```python
##Merge the Data
orders_users_df = pd.merge(
    orders_df,
    users_df,
    on="user_id",
    how="left"
)

final_df = pd.merge(
    orders_users_df,
    restaurants_df,
    on="restaurant_id",
    how="left"
)
```

```python
final_df.to_csv("final_food_delivery_dataset.csv", index=False)
```

```python
##order trends
final_df.groupby("order_date")["order_id"].count()
```

```python
##user behaviour
final_df.groupby("user_id")["total_amount"].sum()

##final_df.columns
```

```python
## City-wise and cuisine-wise performance
final_df.groupby("cuisine")["total_amount"].sum()
```
